```python
import pandas as pd
import numpy as np
```

```python
#download data
com_spnd = pd.read_excel("comspnd.xlsx")
pctfemale = pd.read_excel("womenonbd.xlsx")
```

```python
#drop unnecessary columns
com_spnd.drop(['Ticker', 'Data State', "Field Mnemonic", "Field ID", "2022"], axis = 1, inplace = True)
```

```python
#drop raws containing missing values in each columns 
com_spnd_cleaned = com_spnd.dropna(subset = ["2021","2020","2019","2018"]).reset_index(drop = True)
```

```python
#description contains space before the letter, so make sure to eliminate it

com_spnd_cleaned["Description"] = com_spnd_cleaned["Description"].str.replace(" ", "").astype("str")

#find Media, Apparel/Textile prods, Automotive, Homebuilders, Travel & Lodging, Consumer Products, Oil, Gas & Coal, 
#Asset Management, Life Insurance, Real Estate, 
```

```python
com_spnd_cleaned
```

```python
#do the same process to women on board dataframe
pctfemale.drop(['Ticker', 'Data State', "Field Mnemonic", "Field ID", "2022"], axis = 1, inplace = True)
```

```python
pctfemale_cleaned = pctfemale.dropna(subset = ["2021","2020","2019","2018"]).reset_index(drop = True)
```

```python
pctfemale_cleaned["Description"] = pctfemale_cleaned["Description"].str.replace(" ", "").astype("str")
```

```python
pctfemale_cleaned
```

```python
#now make sure that two dataset has the same company with no missing data
#for the last four years and drop duplicated values
mask = pctfemale_cleaned["Description"].isin(com_spnd_cleaned["Description"])
pctfemaleData = pctfemale_cleaned.loc[mask].reset_index(drop = True).drop_duplicates(subset = "Description", 
                                                                                    keep = "last", 
                                                                                    ignore_index = True)
```

```python
pctfemaleData["Description"].duplicated().sum()
```

```python
#do the samething for community spending
mask = com_spnd_cleaned["Description"].isin(pctfemaleData["Description"])
com_spndData = com_spnd_cleaned.loc[mask].reset_index(drop = True).drop_duplicates(subset = "Description", 
                                                                                    keep = "last", 
                                                                                    ignore_index = True)
```

```python
com_spndData["Description"].duplicated().sum()
```
